```python
class Pokemon:
    def __init__(self, name, level, health, max_health, type, is_ko):
        self.name = name
        self.level = level
        self.health = health
        self.max_health = max_health
        self.type = type
        self.is_ko = is_ko
        
    def __repr__(self):
        return self.name

    def lose_health(self, damage):
        self.health -= damage
        if self.health > 0:
            print("{n} now has {h} health!".format(n=self.name, h=self.health))
        if self.health <= 0:
            self.health = 0
            self.is_ko = True
            print("{n} now has {h} health!".format(n=self.name, h=self.health))
            print("{n} was knocked out!".format(n=self.name))
    def gain_health(self, item):
        if self.is_ko == True and item.restore_ko == False:
            print("{n} is knocked out! You must revive them before using a potion!".format(n=self.name))
        elif self.is_ko == True and item.restore_ko == True:
            print("{n} has been revived!".format(n=self.name))
            self.health += item.health_points
            print("{n}'s health rose by {p}!".format(n=self.name, p=item.health_points))
            print("{n}'s health is now at {h}!".format(n=self.name, h=self.health))
        elif self.health < self.max_health:
                new_health = self.health + item.health_points
                if new_health < self.max_health:
                    self.health = new_health
                    print("{n}'s health rose by {p}!".format(n=self.name, p=item.health_points))
                elif new_health >= self.max_health:
                    print("{n}'s health rose by {p}!".format(n=self.name, p=(self.max_health-self.health)))
                    print("{n} is now at max health!".format(n=self.name))
                    self.health = self.max_health
        elif self.health == self.max_health:
                print("{n}'s health is already maxed out!".format(n=self.name))
    
    def attack(self, defending_pokemon):
        print("{n} attacked {d}!".format(n=self.name, d=defending_pokemon.name))
        if defending_pokemon.type == strong_against.get(self.type):
            defending_pokemon.lose_health(self.level*2)
        elif defending_pokemon.type == weak_against.get(self.type):
            defending_pokemon.lose_health(self.level/2)
        else:
            defending_pokemon.lose_health(self.level)
    
class Items:
    def __init__(self, name, health_points, restore_ko):
        self.name = name
        self.health_points = health_points
        self.restore_ko = restore_ko
        
    def __repr__(self):
        return self.name

strong_against = {"water":"fire", "fire":"grass", "grass":"water"}
weak_against = {"fire":"water", "water":"grass", "grass":"fire"}
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
charmander = Pokemon("Charmander", 3, 50, 50, "fire", False)
squirtle = Pokemon("Squirtle", 3, 50, 50, "water", False)
squirtle2 = Pokemon("Squirtle 2", 3, 50, 50, "water", False)
potion = Items("Potion", 10, False)
revive = Items("Revive", 10, True)
charmander.attack(squirtle)
squirtle.gain_health(revive)
squirtle2.attack(squirtle)
```
